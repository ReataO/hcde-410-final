## Introduction and Motivation ##
Food accessibility is a fundamental aspect of public health. It is a key factor in a population's dietary choices and overall health outcomes. However, systemic inequalities have created disparities in access. Areas affected by these disparities are called “Food Deserts.” These food deserts disproportionately affect low-income households, BIPOC communities, and those with limited transportation options. This project explores potential relationships between demographic & socioeconomic factors, vehicle access, and distance from supermarkets to uncover patterns and potential inequalities in food accessibility across King County. ## Background and Related Work
Previous research has shown that food deserts are prevalent in neighborhoods with high poverty rates and with a significant minority population. Studies highlight that limited access to fresh, nutritious foods in food deserts exacerbates health disparities leading to high rates of obesity, diabetes, and other illnesses. 
- **[Evaluating Consumer Nutrition Environment in Food Deserts and Food Swamps](https://pmc.ncbi.nlm.nih.gov/articles/PMC7967537/)**: This article examines the barriers to food access, particularly in low-income urban neighborhoods, highlighting that a lack of supermarkets and fresh food sources is linked to poorer health outcome
- **[Food Accessibility, Insecurity and Health Outcomes](https://www.nimhd.nih.gov/resources/understanding-health-disparities/food-accessibility-insecurity-and-health-outcomes.html)**: Links food accessibility and insecurity to health disparities, showing that food deserts and low food accessibility lead to poorer health outcomes.
- **[Neighborhood Disparities in Access to Healthy Foods and Their Effects on Environmental Justice](https://pmc.ncbi.nlm.nih.gov/articles/PMC3482049/)**: Analyzes the role of social and environmental factors in food deserts, finding that socioeconomic status, transportation, and community resources are critical in shaping access to healthy foods.
- **[The Intersection of Neighborhood Racial Segregation, Poverty, and Urbanicity and its Impact on Food Store Availability in the United States](https://pmc.ncbi.nlm.nih.gov/articles/PMC3970577/)**: Concludes that poor predominantly black neighborhoods face a double jeopardy with the most limited access to quality food.

The background research supports the focus on demographic factors as a significant contributor to food access disparities. This project aims to add to existing research by examining demographic variables and their roles in food accessibility.

## Findings

### Finding 1: Racial Disparities in Vehicle Access as a Proxy for Food Accessibility
 
#### Code Overview:
This scatterplot illustrates the correlation between six racial demographic groups and the housing units without vehicle count beyond 1/2 mile from a supermarket, the lack of vehicle access is a critical barrier to food accessibility. Each plot quantifies the relationship between race and vehicle access using the Pearson correlation coefficient (r).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'seattle page.csv'
data = pd.read_csv(file_path)

# Create a 2x3 grid for six charts
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # Adjust the size as needed
fig.suptitle("Correlation Analysis: Race  vs Vehicle Access for Food Access", fontsize=16)  # Add a figure-level title

# Data for each chart
charts = [
    {'x': 'TractHispanic', 'y': 'lahunvhalf', 'title': 'Figure 1: Hispanic Housing Units vs Food Access'},
    {'x': 'TractWhite', 'y': 'lahunvhalf', 'title': 'Figure 2: White Housing Units vs Food Access'},
    {'x': 'TractBlack', 'y': 'lahunvhalf', 'title': 'Figure 3: Black Housing Units vs Food Access'},
    {'x': 'TractAsian', 'y': 'lahunvhalf', 'title': 'Figure 4: Asian Housing Units vs Food Access'},
    {'x': 'TractAIAN', 'y': 'lahunvhalf', 'title': 'Figure 5: American Indian vs Food Access'},
    {'x': 'TractOMultir', 'y': 'lahunvhalf', 'title': 'Figure 6: Two or More Races vs Food Access'}
]

# Plot each chart
for idx, chart in enumerate(charts):
    row, col = divmod(idx, 3)  # Get row and column index for the grid
    ax = axes[row, col]  # Access the correct subplot
    
    # Select relevant columns for analysis
    columns_vehicle = [chart['x'], chart['y']]
    vehicle_access = data[columns_vehicle]
    
    # Correlation
    correlation = vehicle_access.corr().iloc[0, 1]
    subtitle = f"Correlation: {correlation:.2f}"
    
    # Scatterplot
    sns.scatterplot(x=chart['x'], y=chart['y'], data=vehicle_access, ax=ax)
    ax.set_title(chart['title'], fontsize=12)
    ax.set_xlabel(chart['x'])
    ax.set_ylabel(chart['y'])
    ax.annotate(subtitle, xy=(0.5, 0.9), xycoords='axes fraction', fontsize=10, ha='center', color='red')  # Add correlation as annotation

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the figure title
plt.show()
```

#### Key Observation:
1. **Hispanic Communities:** This plot has a moderate positive correlation of 0.32, highlighting a significant relationship between Hispanic-majority housing units and limited vehicle access to food. This reinforces existing evidence that Hispanic communities face systemic challenges related to food accessibility.
2. **White Communities:** This plot has a negligible negative correlation of -0.06, indicating an absence of systemic barriers linking White-majority housing units to limited vehicle access. White communities appear largely unaffected by this metric of food accessibility.
3. **Black Communities:** This plot has a moderate positive correlation of 0.28, this correlation underscored that Black-majority housing units experience challenges of limited vehicle access, reinforcing existing evidence of food deserts disproportionately affecting Black communities.
4. **Asain Communities:** With a correlation coefficient of 0.12, there is a weak positive correlation that implies a level of disparity with food accessibility in Asain-majority housing units, although less pronounced than other BIPOC groups.
5. **American Indian/Alaska Native Communities:** This plot has a moderate positive correlation of 0.26, which suggests that Indigenous communities face systemic barriers to food accessibility, this is likely linked to reservation isolation and socioeconomic inequities.
6. **Multiracial Communities:** A moderate positive correlation of 0.30 reveals significant disparities in food accessibility for Multiracial communities, similar to other BIPOC groups.
#### Interpretation:
The Findings indicate a clear pattern: BIPOC communities– including Hispanic, Black, Indigenous, and Multiracial groups—experience higher correlations with limited vehicle access to food compared to predominantly White communities. This is a clear indication that these communities face structural inequities such as the placement of food resources, often concentrated in communities known as food deserts; transportation infrastructure, such as limited availability of public transportation and walkable infrastructure; and the lasting legacy of discriminatory urban planning policies link redlining, which historically limited access to resources and opportunities to BIPOC communities.

The negative correlation for White communities (r = -0.06) is striking and emphasizes the high accessibility that these areas have. Meanwhile, BIPOC communities face compounding challenges, as vehicle access is critical in accessing affordable, healthy food options.

#### Relation to the Research Question:
The analysis strongly supports the hypothesis that BIPOC communities face disproportionately low food accessibility when compared to White communities. This finding reveals that vehicle access—a key factor in food access—varies significantly across racial groups, further cementing the need for targeted interventions to reduce food access disparities.


### Finding 2: Income-Related Disparities in Food Accessibility Based on Distance from Supermarkets
#### Code Overview:
The scatterplot illustrates the relationship between the population with low access to supermarkets ``LAPOP1_10`` and the population with low income and low access to supermarkets ``LALOWI1_10`` within a specified geographic region. The plot shows a positive correlation with a Pearson correlation coefficient of 0.75, indicating a strong linear relationship between these two variables.
- ``LAPOP1_10``: Represents the total number of individuals with limited access to supermarkets (defined as being farther than one mile in urban areas or ten miles in rural areas). 
- ``LALOWI1_10``: Represents the number of individuals who are both low-income and have limited access to supermarkets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'seattle page.csv'
data = pd.read_csv(file_path)

# Select the relevant columns
col1 = 'LAPOP1_10'
col2 = 'LALOWI1_10'

# Drop rows with missing values in the selected columns
data = data.dropna(subset=[col1, col2])

# Convert columns to numeric if necessary
data[col1] = pd.to_numeric(data[col1], errors='coerce')
data[col2] = pd.to_numeric(data[col2], errors='coerce')

# Drop rows with non-numeric values (if any were coerced to NaN)
data = data.dropna(subset=[col1, col2])

# Plot the data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data[col1], y=data[col2])
plt.title('Correlation between LAPOP1_10 and LALOWI1_10')
plt.xlabel('LAPOP1_10')
plt.ylabel('LALOWI1_10')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation = data[col1].corr(data[col2])
print(f'Correlation coefficient between {col1} and {col2}: {correlation:.2f}')
```

#### Key Observations:
- **String positive Correlation:** The data has a strong positive correlation of 0.75, this indicates that as the total population with limited access to supermarkets grows, so does the subpopulation with low income and limited access. This finding supposes the hypothesis that income-related disparities are closely tied to barriers to accessing food– the greater distances exacerbate income-related disparities.
- **Concentration of Data at Lower Ranges:** The cluster of data points in the lower range of both aces indicates that, in many communities, the absolute number of individuals with limited access remains relatively modest. However, outliers with substantially higher values in both variables. This indicates that some communities experience acute disparities, with large numbers of low-income individuals disproportionately affected by limited access to supermarkets. 

#### Implications:
The visualization demonstrates that the further one lives from supermarkets is a significant determinant of food accessibility, particularly for those who belong to low-income populations. Furthermore, the strong correlation suggests that income is an amplification, specifically in regions where the general population experiences difficulty in accessing food due to their distance from a supermarket. 

#### Relation to the Research Question:
This analysis contributes directly to understanding how distance from supermarkets and income-related factors intersect to create barriers to food accessibility. The strong correlation between the two variables LAPOP1_10 and LALOWI1_10 highlights that income is a critical factor in determining the degree to which limited access to supermarkets translates into food insecurity.
